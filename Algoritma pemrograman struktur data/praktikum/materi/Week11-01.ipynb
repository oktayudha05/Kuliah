<h1><b>Python Modules</b></h1>
What is a Module?
Consider a module to be the same as a code library.

A file containing a set of functions you want to include in your application.

<h2><b>Create a Module</b></h2>
To create a module just save the code you want in a file with the file extension .py:

Save this code in a file named mymodule.py
<code><p>
def greeting(name):

  print("Hello, " + name)
</p></code>

<h2><b>Use a Module</b></h2>
Now we can use the module we just created, by using the import statement:

**Example :**


```python
#Import the module named mymodule, and call the greeting function:

import mymodule

mymodule.greeting("Namamu")
```

<b>Note:</b> When using a function from a module, use the syntax: module_name.function_name.

<h2><b>Variables in Module</b></h2>
The module can contain functions, as already described, but also variables of all types (arrays, dictionaries, objects etc):


Save this code in the file mymodule2.py
<p>
<code>
person1 = {
  
  "name": "Brian",
  
  "age": 36,
  
  "country": "Indonesia"

}
</code>
</p>

**Example :**

```python
#Import the module named mymodule, and access the person1 dictionary:

import mymodule2

a = mymodule2.person1["age"]
print(a)
```

<h2><b>Naming a Module</b></h2>
You can name the module file whatever you like, but it must have the file extension <code>.py</code>
<h2><b>Re-naming a Module</b></h2>
You can create an alias when you import a module, by using the as keyword:

**Example :**

```python
#Create an alias for mymodule called mx:

import mymodule2 as mx

a = mx.person1["age"]
print(a)
```

<h2><b>Built-in Modules</b></h2>
There are several built-in modules in Python, which you can import whenever you like.

**Example :**


```python
#Import and use the platform module:

import platform

x = platform.system()
print(x)
```

<h2><b>Using the dir() Function</b></h2>
There is a built-in function to list all the function names (or variable names) in a module. The <cod>dir()</code> function:

**Example :**


```python
#List all the defined names belonging to the platform module:

import platform

x = dir(platform)
print(x)
```

<b>Note:</b> The dir() function can be used on all modules, also the ones you create yourself.

<h2><b>Import From Module</b></h2>
You can choose to import only parts from a module, by using the from keyword.

**Example :**


The module named mymodule3.py has one function and one dictionary:



<code>
def greeting(name):

  print("Hello, " + name)


person1 = {

  "name": "namamu",

  "age": 36,

  "country": "Indonesia"

}
</code>


**Example :**
Import only the person1 dictionary from the module:



```python
from mymodule3 import person1

print (person1["age"])
```

**Note:** When importing using the from keyword, do not use the module name when referring to elements in the module. Example: <code>person1["age"]</code>, not <code>mymodule3.person1["age"]</code>
